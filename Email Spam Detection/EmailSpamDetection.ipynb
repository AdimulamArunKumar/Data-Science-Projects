```python
# Import necessary modules and libraries
import pandas as pd
```

```python
#Load the dataset

data = pd.read_csv("C:\\Users\\arunk\\OneDrive\\Documents\\spam.csv", encoding='ISO-8859-1') # The varirable "encoding='ISO-8859-1'" is used to fix Unicode errors.
```

```python
data.head() # displays the first 5 rows of the dataset.
```

Removing Unnecessary Columns

```python
# Removing Unnecessary Columns

for col in data.columns:
    if 'Unnamed' in col:
        del data[col]
data.head()
```

Renaming Columns

```python
# Renaming Columns

data.columns = ['Category','Message']
data.head()     
```

Training, Testing and Splitting the data using scikit-learn

```python
# Training, Testing and Splitting the data using scikit-learn

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data.Message,data.Category,test_size=0.2)
```

Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
x_train_count = v.fit_transform(x_train.values)
x_train_count.toarray()[:2]
features = v.fit_transform(x_train)

print(x_train_count)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
     
```

```python
features_test = v.transform(x_test)
model.score(features_test,y_test)*100
```
